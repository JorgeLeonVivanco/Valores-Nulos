```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el archivo CSV
csv_path = "Barcelona_Filtrado.csv"  # Asegurar que esté en el mismo directorio de ejecución
df = pd.read_csv(csv_path)
df
```

```python
# Identificar valores nulos
valores_nulos = df.isnull().sum()
columnas_con_nulos = valores_nulos[valores_nulos > 0]
print("Columnas con valores nulos:\n", columnas_con_nulos)
```

```python
df_limpio = df.copy()
```

```python
# Tratamiento de valores nulos por columna
# 1. Promedio (Media)
df_limpio['review_scores_cleanliness'] = df_limpio['review_scores_cleanliness'].fillna(df_limpio['review_scores_cleanliness'].mean())
df_limpio['review_scores_checkin'] = df_limpio['review_scores_checkin'].fillna(df_limpio['review_scores_checkin'].mean())
df_limpio['review_scores_communication'] = df_limpio['review_scores_communication'].fillna(df_limpio['review_scores_communication'].mean())
df_limpio['review_scores_rating'] = df_limpio['review_scores_rating'].fillna(df_limpio['review_scores_rating'].mean())
df_limpio['review_scores_value'] = df_limpio['review_scores_value'].fillna(df_limpio['review_scores_value'].mean())
df_limpio['review_scores_location'] = df_limpio['review_scores_location'].fillna(df_limpio['review_scores_location'].mean())
df_limpio['review_scores_accuracy'] = df_limpio['review_scores_accuracy'].fillna(df_limpio['review_scores_accuracy'].mean())
```

```python
# 2.Mediana
# Convert 'price' column to numeric before calculating the median
df_limpio['price'] = pd.to_numeric(df_limpio['price'].str.replace('[\$,]', '', regex=True), errors='coerce')
df_limpio['price'] = df_limpio['price'].fillna(df_limpio['price'].median())
df_limpio['beds'] = df_limpio['beds'].fillna(df_limpio['beds'].median())
df_limpio['bedrooms'] = df_limpio['bedrooms'].fillna(df_limpio['bedrooms'].median())
```

```python
# 3.Sustitución con un valor específico
df_limpio['bathrooms_text'] = df_limpio['bathrooms_text'].fillna('1 bath')
df_limpio['host_total_listings_count'] = df_limpio['host_total_listings_count'].fillna(1)
df_limpio['host_listings_count'] = df_limpio['host_listings_count'].fillna(1)
df_limpio['has_availability'] = df_limpio['has_availability'].fillna('unknown')
```

```python
# 4.Sustitución con "Desconocido" en variables categóricas
df_limpio['host_name'] = df_limpio['host_name'].fillna('Desconocido')
df_limpio['host_neighbourhood'] = df_limpio['host_neighbourhood'].fillna('Desconocido')
df_limpio['neighbourhood'] = df_limpio['neighbourhood'].fillna('Desconocido')
df_limpio['neighborhood_overview'] = df_limpio['neighborhood_overview'].fillna('Desconocido')
df_limpio['host_location'] = df_limpio['host_location'].fillna('Desconocido')
df_limpio['description'] = df_limpio['description'].fillna('No description available')
```

```python
# 5.Sustitución con forward fill (ffill)
df_limpio['host_acceptance_rate'] = df_limpio['host_acceptance_rate'].ffill()
df_limpio['host_response_rate'] = df_limpio['host_response_rate'].ffill()
df_limpio['host_response_time'] = df_limpio['host_response_time'].ffill()
```

```python
# 6. **Sustitución con backward fill (bfill)**
df_limpio['host_identity_verified'] = df_limpio['host_identity_verified'].bfill()
df_limpio['host_is_superhost'] = df_limpio['host_is_superhost'].bfill()
df_limpio['host_verifications'] = df_limpio['host_verifications'].bfill()
```

```python
# 7. **Sustitución con la fecha más frecuente**
df_limpio['first_review'] = df_limpio['first_review'].fillna(df_limpio['first_review'].mode()[0])
df_limpio['last_review'] = df_limpio['last_review'].fillna(df_limpio['last_review'].mode()[0])
```

```python
# **Verificar que no haya valores nulos después de la limpieza**
valores_nulos_post = df_limpio.isnull().sum()
print("\nValores nulos por columna después del tratamiento:")
print(valores_nulos_post[valores_nulos_post > 0])

total_nulos_post = valores_nulos_post.sum()
print("\nTotal de valores nulos después del tratamiento:", total_nulos_post)
```

```python
# Guardar el archivo limpio
df_limpio.to_csv("Barcelona_Filtrado_Limpio.csv", index=False)
```
