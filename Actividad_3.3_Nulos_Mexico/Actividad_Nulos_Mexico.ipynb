```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Mexico_50_Columnas.csv", on_bad_lines='skip', quoting=1, encoding='utf-8')
df
```

```python
# Mostrar primeras filas del dataset
df.head()
```

```python
# Identificar valores nulos por columna
valores_nulos = df.isnull().sum()
print("Valores nulos por columna:")
print(valores_nulos)
```

```python
# Método de imputación de valores nulos
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Para valores categóricos, reemplazar nulos con la moda usando mapeo de diccionario
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Manejo de valores nulos en la columna 'neighbourhood_group_cleansed'
if df['neighbourhood_group_cleansed'].isnull().sum() > 0:
    df['neighbourhood_group_cleansed'].fillna("Indefinido", inplace=True)
```

```python
# Verificar que no queden valores nulos
print("Valores nulos después de la imputación:")
print(df.isnull().sum())
```

```python
# Guardar el dataset limpio
df.to_csv('Mexico_Limpio_SinNulos.csv')
```
